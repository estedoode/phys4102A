### Next, I will use a Gaussian Fit Function to fit two spectral peaks presented in each of the chosen plots for the purpose of obtaining the a linear scale function to convert each plots time scale into wavelength. This way the finalized plots for my report will show wavelength as means to narrow down the corresponding wavelength for each spectral line. 

### The Gaussian function used for each of the fits is as follows: 

$ f(x) = a e^{-\frac{(x-b)^{2}}{2c^{2}}} $ 

### where, a = height of the curve's peak, b = the position of the center of the peak, c = standard deviation, e = Euler's number, x = integer

### I selected the Hydrogen Trial 1 data to carry out the rest of the data analysis. The NIST wavelengths that correspond to the two peaks that will be analyzed are 486 nm and 656 nm. 

### The purpose of this data analysis is to find the corresponding time values of the peaks. This way, both the $\Delta \lambda$ and $\Delta t$ are known allowing for the slope to be calculated.  

## Hydrogen Gaussian Fit for time scale, returns $t_1$ which corresponds with $\lambda$ = 486.1 nm

```python
# hydrogen1 trial 1 data  
rHydrogen1 = range(0,len(Hydrogen1)) # Range of the hydrogen data 
xhydro_t1 = rHydrogen1[3308:3325]
yhydro_t1 = Hydrogen1[3308:3325]

# Gaussian Fit Function 
def Gauss(x,a,b,c): # b equals the mean 
    return a * np.exp(-(x-b)**2/(2*c**2))

p_hydro1T1 = [6.9,3314,.25] # 3.65
hydroParam1_trial1, hydroErr1_trial1 = optimize.curve_fit(Gauss,xhydro_t1,yhydro_t1,p0 =p_hydro1T1) #[6.5,80,2000])
hydroT1fit1 = Gauss(xhydro_t1,*hydroParam1_trial1)

plt.figure(figsize = (8,6))
plt.plot(rHydrogen1,Hydrogen1,'.',color = 'k')
plt.plot(xhydro_t1,hydroT1fit1,'r',label='Gaussian_Fit')
plt.title('Hydrogen Trial 1 Gaussian Fit at 486 nm',fontsize = 14)
plt.xlabel('Time',fontsize = 14)
plt.ylabel('Voltage (V)',fontsize = 14)
plt.xlim([3308,3325])
plt.show()
print('Parameter Values (a,b,c) = ',hydroParam1_trial1)
print('\n',np.sqrt(hydroErr1_trial1))
```

## Hydrogen Gaussian Fit for time scale, returns $t_2$ which corresponds with $\lambda$ = 656.6 nm

```python
# 2nd Gaussian Fit 
xhydro2_t1 = rHydrogen1[8765:8776]
yhydro2_t1 = Hydrogen1[8765:8776] 

p_hydro2T1 = [2,50,3700]
hydroParam2_trial1, hydroErr2_trial1 = optimize.curve_fit(Gauss,xhydro2_t1,yhydro2_t1,p0 =p_hydro2T1)
hydroT1fit2 = Gauss(xhydro2_t1,*hydroParam2_trial1)

plt.figure(figsize = (8,6))
plt.plot(rHydrogen1,Hydrogen1,'.',color = 'k')
plt.plot(xhydro2_t1,hydroT1fit2,'r',label='Gaussian_Fit')
plt.title('Hydrogen Trial 1 Gaussian Fit at 656 nm peak',fontsize = 14)
plt.xlabel('Time',fontsize = 12)
plt.ylabel('Voltage (V)',fontsize = 12)
plt.xlim([8765,8775])
plt.show()
print('Parameter Values (a,b,c) = ',hydroParam2_trial1)
print('\n',np.sqrt(hydroErr2_trial1))
```

### Conversion of the time scale to the spectral lines 

### From the Gaussian Fits the mean time was obtained which allowed for the calculation of the slope and y-intercept. 

### In the two fits, the wavelength are known and the mean time is found which can be plugged into the slope equation. Using formula $\lambda(t) = mt+b$, this will convert the time scale to the wavelength. I will show the process done for hydrogen. All the values used for these calculations come from the gaussian fits and us knowing the corresponding wavelength of each spectral peak. 

## Given: $t_1$ = 3315.21, $t_2$ = 8770.39, $\lambda_1$ = 486.1 nm, $\lambda_2$ = 656.3 nm

## $\lambda (t) = mt + b $ 
## $ m = \frac{\Delta\lambda}{\Delta t} = \frac{656.3 ~nm - 486.1 ~nm}{8770.39 ~ds - 3315.21 ~ds} = 3.120 x 10^{-2} ~\frac{nm}{ds}$
## $ b = \lambda_2 - mt_2 = 656.3 ~nm - (3.120 x 10^{-2} ~\frac{nm}{ds}) (8770.39 ~ds) = 382.66 ~nm $ 


## The slope and y-intercept of the linear scale function used to convert the time scale to a wavelength scale experience error. Since the measurements depend on the slope and y-intercept, then the propagation of errors needs to be used. Down below is the error for slope and the y-interept of hydrogen. 


## For Hydrogen


## Need to find uncertainty for change in time. We can calculate this using propagation of errors method. 

## $ \delta_{\Delta t} = \sqrt{(\delta t_1)^2 + (\delta t_2)^2} $ , where $\delta t_1$ and $\delta t_2$ are given by the parameter "c" of the Gaussian function which corresponds with the standard deviation. 

## $\delta_{t_1}$ = 3.65 ds , $ \delta t_2$ = 3.64 ds 

## $\delta_{\Delta t}$ = 5.15 

## This can be now used to calculate the error of the slope

## $\delta_{ m} = \sqrt{(\frac{\delta_{ \Delta \lambda}}{\Delta t})^2 + (\frac{\Delta \lambda \cdot \delta_ {\Delta t}}{\Delta t^2})^2}$

## $\delta_m = 2.95 \cdot 10^{-5} ~\frac{nm}{ds}$ 

## Now we calculate error of the y-intercept b applying the propagation of error as well. 

## $\delta_b = \sqrt{(\frac{\partial b}{\partial m} \delta_m)^2+(\frac{\partial b}{\partial t_2}\delta_{t_2})^2} $ 

## Reduces to 

## $\delta_b = \sqrt{(t_2 \delta_m)^2 + (m\delta_{t_2})^2} =.28~nm $ 


## Result: $\lambda (t) = (3.120 \pm .002) \cdot 10^{-2} + (382.7 \pm .3)~nm $ 

### Plugging values m and b into the next box which contains a function called linfun, it converts the time scale to wavelength scale. As a result, the graph shows the hydrogen emission spectra (Voltage vs Wavelength) 

```python
# Convert the time to wavelength 

def linfun(t,m,b):
    return m*t + b
arrHydrogen1 = np.arange(0,len(Hydrogen1))

hydro_wavelength = linfun(arrHydrogen1,3.120e-2,382.7)

plt.figure(figsize = (8,6))
plt.plot(hydro_wavelength,Hydrogen1,color = 'k')#label = 'Hydrogen Spectral Lines')
plt.title('Hydrogen Emission Spectral Lines',fontsize = 14)
plt.xlabel('Wavelength (nm)',fontsize  = 12)
plt.ylabel('Voltage (V)', fontsize = 12)
#plt.ylim(0,9)
#plt.legend(loc = 'upper right',fontsize = 12)

#plt.xlim(hydro_wavelength[8771],hydro_wavelength[9000])

plt.show()
```

# -----------------------------------------------------------------------------------------------------------
### Converting Neon Time Scale to Wavelength

## $ t_1 = 7011.68~ds $ corresponds with $\lambda_1 = 607.4~nm$  
## $ t_2 = 8063.57~ds $ corresponds with $\lambda_2 = 640.2~nm $ 
## $\lambda (t) = mt + b $ 
## Calculation for slope: 
## $ m = \frac{\Delta\lambda}{\Delta t}=\frac{\lambda_2-\lambda_1}{t_2-t_2} = \frac{640.2 ~nm - 607.4~ nm}{8063.57~ ds - 7011.68~ ds} = 3.118 x 10^{-2}~ \frac{nm}{ds}$
## Calculation for y-intercept: 
## $ b = \lambda_2 - mt_2 = 640.2~nm - (3.118 x 10^{-2}~\frac{nm}{ds})(8063.57~ds) = 388.78 ~ nm $ 

## Need to find uncertainty for change in time. We can calculate this using propagation of errors method. 

## $ \delta_{\Delta t} = \sqrt{(\delta t_1)^2 + (\delta t_2)^2} $ , where $\delta t_1$ and $\delta t_2$ are given by the parameter "c" of the Gaussian function which corresponds with the standard deviation. 

## $\delta_{t_1}$ = 3.84 ds , $ \delta t_2$ = 3.67 ds 

## $\delta_{\Delta t}$ = 5.31 ds  

## This can be now used to calculate the error of the slope

## $\delta_{ m} = \sqrt{(\frac{\delta_{ \Delta \lambda}}{\Delta t})^2 + (\frac{\Delta \lambda \cdot \delta_ {\Delta t}}{\Delta t^2})^2}$

## $\delta_m = 1.644 \cdot 10^{-4} ~\frac{nm}{ds}$ 

## Now we calculate error of the y-intercept b applying the propagation of error as well. 

## $\delta_b = \sqrt{(\frac{\partial b}{\partial m} \delta_m)^2+(\frac{\partial b}{\partial t_2}\delta_{t_2})^2} $ 

## Reduces to 

## $\delta_b = \sqrt{(t_2 \delta_m)^2 + (m\delta_{t_2})^2} =1.3327~nm $ 


## Result: $\lambda (t) = (3.118 \pm 1.644) \cdot 10^{-2} + (388.8 \pm 1.3)~nm $
